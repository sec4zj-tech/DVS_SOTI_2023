```python
# Replace 'df' with the actual name of your pandas dataframe
unfinished_count = 0

# Iterate over each row in the dataframe
for index, row in df.iterrows():
    # Check if the string "unfinished" is present in any of the values in the row
    if row.str.contains('unfinished').any():
        unfinished_count += 1

print(f"Number of rows containing 'unfinished': {unfinished_count}")
```

```python
# Replace 'df' with the actual name of your pandas dataframe
indexes_to_drop = []

# Iterate over each row in the dataframe
for index, row in df.iterrows():
    # Check if the string "unfinished" is present in any of the values in the row
    if row.str.contains('unfinished').any():
        indexes_to_drop.append(index)

# Drop the rows with indexes in 'indexes_to_drop'
df = df.drop(indexes_to_drop)

# Reset the index to avoid gaps after dropping rows
df = df.reset_index(drop=True)
```
